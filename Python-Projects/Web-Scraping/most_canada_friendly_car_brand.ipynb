```python
import pandas as pd
import requests


url = "https://github.com/fababa/Data-Analyst-Portfolio---IBM-Data-Analyst-and-Azure-AI-Fundamentals/blob/ef3966a635689b6639dfeec8d9f67f64a15471df/Python-Projects/Web-Scraping/most_canada_friendly_car_brand.json"


response = requests.get(url)

if response.status_code == 200:
  try:
    data = response.json()
    post_data = data[0]["data"]["children"][0]["data"]
    comments = []
    def extract_comments(comment_list):
      for comment in comment_list:
        if comment["kind"] == "t1":
          comment_data = comment["data"]
          comments.append({
              "post_title": post_data.get("title"),
              "post_author": post_data.get("author"),
              "comment_author": comment_data.get("author"),
              "comment": comment_data.get("body"),
              "score": comment_data.get("score"),
              "created_utc": comment_data.get("created_utc")
              })
          replies = comment_data.get("replies")
          if isinstance(replies, dict):
            extract_comments(replies["data"]["children"])
            extract_comments(data[1]["data"]["children"])
            comments_df = pd.DataFrame(comments)
            print(comments_df.head())
            comments_df.to_csv("reddit_canada_friendly_car_brand_comments.csv", index=False)
            print("Scraping completed and CSV file saved.")
  except ValueError:
        print("Response is not valid JSON.")
        print(response.text[:500])

else:
    print("Failed to scrape Reddit page.")
    print("Status code:", response.status_code)
```
