```python
#Q17 Create the following dictionary encryption={10:"hungry", 101:"am", 110:"I",
# What should be the decryption of the following code "110 101 10 1011 110 1001 1
encryption = {10:"hungry", 101:"am", 110:"I"}
encryption_keys = encryption.keys();
key_to_search_tuple = (110, 101, 10, 1011, 110, 1001, 1)
decoded_str = ""
for key_to_search in key_to_search_tuple:
    for key_decryption in encryption_keys:
        if (key_decryption == key_to_search):
            decoded_str = decoded_str + " " + encryption[key_decryption]
        else:
            pass


print(decoded_str)
```
